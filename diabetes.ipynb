#Correlation & Feature Engineering

```python
#Emphasizes numeric feature correlations to identify relationships such as age and BP.
plt.figure(figsize=(12, 8))
n_df = df.select_dtypes(include='number')
plt.figure(figsize=(12, 8))
sns.heatmap(n_df.corr(), annot=True, cmap='cividis', fmt=".2f", square=True)
plt.title("Correlation Heatmap of Numerical Features")
```

#Population Profiling

```python
#Provides the distribution of BMI to determine the weight categories impacting diabetes risk.
plt.figure(figsize=(10, 6))
sns.histplot(df["BMI"], bins=60, kde=True, edgecolor = 'white', color='black')
plt.title("Distribution of BMI")
plt.xlabel("BMI")
```

```python
#Displays physical activity levels in the population to understand lifestyle patterns.
ac = df["Physical_Activity_Level"].value_counts().reset_index() #activity count
ac.columns = ["Physical_Activity_Level", "Count"]
plt.figure(figsize=(10, 6))
sns.barplot(x="Physical_Activity_Level", y="Count", data=ac, palette="Set2")
plt.title("Distribution of Physical Activity Levels in the Dataset")
plt.xlabel("Physical Activity Level")
plt.ylabel("Number of Individuals")
plt.xticks(rotation=45)
```

```python
#Compares mean fasting glucose by smoking status to demonstrate that smokers can have higher glucose.
plt.figure(figsize=(8, 6))
avg_glucose = df.groupby("Smoking_Status")["Fasting_Blood_Glucose"].mean().reset_index()
sns.barplot(x="Smoking_Status", y="Fasting_Blood_Glucose", data=avg_glucose, palette="viridis")
plt.title("Average Fasting Glucose by Smoking Status")
plt.xlabel("Smoking Status")
plt.ylabel("Average Fasting Glucose")
plt.xticks(rotation=0)
```

```python
#Compares BMI distribution between men and women to appreciate gender-based differences in weight.
plt.figure(figsize=(8, 6))
sns.violinplot(x="Sex", y="BMI", data=df, palette="Set3", inner="quartile")
plt.title("BMI Distribution by Sex")
plt.xlabel("Sex")
plt.ylabel("BMI")
```

```python
#Compares HDL and LDL cholesterol by sex for gender-based heart health risk.
plt.figure(figsize=(12,10))
sns.scatterplot(x="Cholesterol_HDL", y="Cholesterol_LDL", data=df, hue="Sex", alpha=0.6)
plt.title("LDL vs HDL by Sex")
plt.xlabel("HDL ")
plt.ylabel("LDL ")
```

#Risk Factor VS Outcomes Relationship

```python
#Indicates an increasing trend with age and HbA1c, where blood sugar rises with age.
plt.figure(figsize=(15, 8))
sns.regplot(x="Age", y="HbA1c", data=df, scatter_kws={"alpha": 0.6}, line_kws={"color": "red"})
plt.grid(axis='y')
plt.title("Age vs. HbA1c with Regression Line")
plt.xlabel("Age")
plt.ylabel("HbA1c")
```

```python
#Compares HbA1c values by activity level, demonstrating that more active subjects have improved blood glucose.
plt.figure(figsize=(10,6))
sns.boxplot(x='Physical_Activity_Level', y='HbA1c', data=df, palette='Set3', showmeans=True)
plt.title('HbA1c by Physical Activity Level')
```

```python
# Examines how the level of physical activity impacts systolic blood pressure.
plt.figure(figsize=(10, 6))
sns.boxplot(x="Physical_Activity_Level", y="Blood_Pressure_Systolic", data=df, palette="Set2", showmeans=True)
plt.title("Systolic Blood Pressure by Physical Activity Level")
plt.xlabel("Physical Activity Level")
plt.ylabel("Systolic Blood Pressure")
plt.xticks(rotation=45)
```

```python
#Pools fasting glucose by smoking status and gender to investigate combined lifestyle effects.
plt.figure(figsize=(12, 6))
sns.violinplot(x='Smoking_Status', y='Fasting_Blood_Glucose', hue='Sex', data=df, split=True, palette='coolwarm')
plt.title('Fasting Glucose by Smoking Status and Sex')
plt.xlabel('Smoking Status')
plt.ylabel('Fasting Glucose')
```

```python
# Show how mean HbA1c rises with age groups, showing increased risk of diabetes in elderly.
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 30, 45, 60, 75, 100], labels=['<30', '30-45', '45-60', '60-75', '75+'])
age_hba1c = df.groupby('Age_Group')['HbA1c'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.lineplot(x='Age_Group', y='HbA1c', data=age_hba1c, marker='o', color='purple')
plt.grid(axis='y')
plt.title("Average HbA1c by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average HbA1c")
```

#Demographic Based Health Analysis

```python
#Examines HbA1c levels across ethnicities to detect disparities in diabetes risk.
plt.figure(figsize=(10, 6))
sns.boxplot(x="Ethnicity", y="HbA1c", data=df, palette="pastel", showmeans=True)
plt.title("HbA1c Distribution by Ethnicity")
plt.xlabel("Ethnicity")
plt.ylabel("HbA1c")
plt.xticks(rotation=45)
```

```python
#Indicates the proportion of people with diabetes in each ethnic group to draw attention to inequalities in diabetes occurrence.
df['Diabetic'] = df['HbA1c'] >= 6.5
ethnicity_diabetes = df.groupby('Ethnicity')['Diabetic'].mean().reset_index()
ethnicity_diabetes['Diabetic'] *= 100
plt.figure(figsize=(10, 6))
sns.barplot(x='Ethnicity', y='Diabetic', data=ethnicity_diabetes, palette='muted')
plt.title('Diabetes Prevalence (%) by Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Percent Diabetic')
plt.xticks(rotation=45)
```

```python
#Contrasts diabetic vs. non-diabetic numbers by gender to observe gender differences in diabetes occurrence.
# Threshold for diabetes is 6.5
df['Diabetic'] = df['HbA1c'] >= 6.5
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Diabetic', data=df, palette='coolwarm')
plt.title("Diabetes Diagnosis Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title='Diabetic')
```

```python
#Displays the percentage of diagnosed diabetes occurrences by ethnicity.
ethnicity_diabetes = df.copy()
ethnicity_diabetes['Diabetic'] = ethnicity_diabetes['HbA1c'] >= 6.5
ethnicity_counts = ethnicity_diabetes[ethnicity_diabetes['Diabetic'] == True]['Ethnicity'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(ethnicity_counts, labels=ethnicity_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Distribution of Diagnosed Diabetes Cases by Ethnicity")
```

```python
#Compares HbA1c levels by sex across ethnicities to show demographic differences in blood sugar levels.
plt.figure(figsize=(12, 6))
sns.boxplot(x='Ethnicity', y='HbA1c', hue='Sex', data=df, palette='Set2')
plt.title('HbA1c Distribution by Sex Across Ethnicities')
plt.xlabel('Ethnicity')
plt.ylabel('HbA1c')
plt.xticks(rotation=45)
plt.legend(title='Sex')
```

```python
#Tracks BMI changes by age group and sex to reveal how weight trends vary with age and gender.
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 30, 45, 60, 75, 100], labels=['<30', '30-45', '45-60', '60-75', '75+'])
bmi_age_sex = df.groupby(['Age_Group', 'Sex'])['BMI'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=bmi_age_sex, x='Age_Group', y='BMI', hue='Sex', marker='o')
plt.title('Average BMI by Age Group and Sex')
plt.xlabel('Age Group')
plt.ylabel('BMI')
```

# **Overall Results**
1. The population has a broad BMI range, with physical inactivity associated with increased blood pressure and HbA1c levels, reflecting a strong association between lifestyle and risk of diabetes.

2. Smokers have higher fasting glucose levels, particularly in males, which implies that smoking can contribute to impaired glucose regulation.

3. HDL and LDL cholesterol distributions differ significantly by sex, illustrating biological risk variations in heart conditions among diabetics.

4. Both HbA1c and BMI increase with age, particularly among men, illustrating age as a dominant risk factor for diabetes and metabolic transformation.

5. Strong differences were seencertain ethnic groups and men have greater diabetes rates, with visualizations affirming that these populations require more specialized interventions.

6. The heatmap of correlations shows strong correlation between age, glucose, blood pressure, and HbA1cupholding their significance for inclusion in diabetes prediction models.
